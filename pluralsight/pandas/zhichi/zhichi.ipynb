```python
cond1 = (df['来源类型'] == '访客')
cond2 = (df['MsgLen'] <= 5)
cond3 = (df['消息内容'].str.contains('^[你您]好'))
cond4 = (df['消息内容'].str.contains('^[^\u4e00-\u9fa5]*$'))
cond5 = (df['消息内容'].str.contains('人工|客服'))
# cond6 = (df['消息内容'].str.contains('电话|热线|座机'))
cond6 = (df['消息内容'].str.contains('有人吗|在吗|不是|谢谢|再见'))
cond7 =  (df['MsgLen'] <= 1)
```

```python
msgdf = df[((cond1&cond2)&(cond3|cond4|cond5|cond6|cond7))].filter(['来源类型', '消息内容'])
msgdf
```

```python
msgdf = df[~((cond1&cond2)&(cond3|cond4|cond5|cond6|cond7))].filter(['来源类型', '消息内容'])
msgdf
```

```python
session_groups = df.groupby('会话ID')
```

```python
flags = {}
for session_id, session_df in session_groups:
#     print(session_id)
#     print('----------------------')
    found = False
    for idx, row in session_df.iterrows():
        if found:
            flags[idx] = 0
            continue
        if row['来源类型'] == '访客' and row['MsgLen'] > 3:
            flags[idx] = 1
            found = True
        else:
            flags[idx] = 0
#         print(type(row['MsgLen']))
#         print(row['来源类型'])
#         print(row['MsgLen'])
#     print(flags)
#     session_df = session_df.assign(IsFirstQ=pd.Series(flags))
#     print(session_df)
df = df.assign(IsFirstQ=pd.Series(flags, dtype='int32'))
df
```
